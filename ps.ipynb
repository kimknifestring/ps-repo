# 백준 2146번 다리 만들기 골드3

```python
from collections import deque
import sys

# 입력 속도 개선
input = sys.stdin.readline

N = int(input())
arr = []
for _ in range(N):
    arr.append(list(map(int, input().split())))

visited = [[False] * N for _ in range(N)]
dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]
island_num = 1

# 1단계: BFS로 각 섬에 고유 번호 붙이기 (2, 3, 4...)
def number_islands(x, y):
    global island_num
    q = deque([(x, y)])
    visited[x][y] = True
    arr[x][y] = island_num
    
    while q:
        cur_x, cur_y = q.popleft()
        for i in range(4):
            nx, ny = cur_x + dx[i], cur_y + dy[i]
            if 0 <= nx < N and 0 <= ny < N and not visited[nx][ny] and arr[nx][ny] == 1:
                visited[nx][ny] = True
                arr[nx][ny] = island_num
                q.append((nx, ny))

for i in range(N):
    for j in range(N):
        if arr[i][j] == 1 and not visited[i][j]:
            number_islands(i, j)
            island_num += 1

# 2단계: "만남의 장소" 방식으로 최단 다리 찾기
dist = [[-1] * N for _ in range(N)]
q = deque()

# 모든 육지 칸을 시작점으로 큐에 추가
for i in range(N):
    for j in range(N):
        if arr[i][j] > 0:
            q.append((i, j))
            dist[i][j] = 0

min_bridge = float('inf')

while q:
    x, y = q.popleft()
    
    for i in range(4):
        nx, ny = x + dx[i], y + dy[i]
        
        if 0 <= nx < N and 0 <= ny < N:
            # 다른 섬의 영토를 만난 경우 (다리 완성)
            if arr[nx][ny] > 0 and arr[nx][ny] != arr[x][y]:
                bridge_len = dist[x][y] + dist[nx][ny]
                min_bridge = min(min_bridge, bridge_len)
            
            # 주인이 없는 바다를 만난 경우 (영토 확장)
            if arr[nx][ny] == 0 and dist[nx][ny] == -1:
                dist[nx][ny] = dist[x][y] + 1
                arr[nx][ny] = arr[x][y] # 바다 채워버려
                q.append((nx, ny))

print(min_bridge)
```

# 백준 10971 외판원 순회 2

```python
from collections import deque
N=int(input())
arr=[]
visited=[False for _ in range(N+1)]
g={}
for _ in range(N):
    line=list(map(int,input().split()))
    arr.append(line)

for i in range(N):
    for j in range(N):
        if arr[i][j]!=0:
            g.setdefault(i,[]).append((j,arr[i][j]))

min_cost=float('inf')
def dfs(start_city,current_city,count,current_cost):
    global min_cost
    if current_cost>=min_cost:
        return
    if count == N:
        if arr[current_city][start_city]!=0:
            total_cost=current_cost+arr[current_city][start_city]
            min_cost=min(min_cost,total_cost)
        return
    for city in g[current_city]:
        if arr[current_city][city[0]]!=0 and not visited[city[0]]:
            visited[city[0]]=True
            dfs(city[0],count+1,current_cost+city[1])
            visited[city[0]]=False

for city in g:
    visited[city]=True
    dfs(city,city,1,0)
    visited[city]=False

print(min_cost)
```

# 백준 1987 알파벳 골드 4

```python
from collections import deque

R, C = map(int, input().split())
arr = [list(input()) for _ in range(R)]

dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]

start = 1 << (ord(arr[0][0]) - ord('A'))

visited = set()
visited.add((0, 0, start))

q = deque()
q.append((0, 0, start, 1))

mxcount = 1

while q:
    x, y, bitmask, count = q.popleft()
    mxcount = max(mxcount, count)

    for i in range(4):
        nx, ny = x + dx[i], y + dy[i]
        if 0 <= nx < C and 0 <= ny < R:
            c = ord(arr[ny][nx]) - ord('A')
            if not (bitmask & (1 << c)):
                nbitmask = bitmask | (1 << c)
                state = (ny, nx, nbitmask)
                if state not in visited:
                    visited.add(state)
                    q.append((nx, ny, nbitmask, count + 1))

print(mxcount)
```

# 케빈 베이컨의 6단계 법칙

```python
from collections import deque
N,M=map(int,input().split())
dic={}
for _ in range(M):
    A,B=map(int,input().split())
    dic.setdefault(A,[]).append(B)
    dic.setdefault(B,[]).append(A)

kevinList=[0 for _ in range(N+1)]

def bfs(idx):
    q=deque([(idx,0)])
    visited=[False for _ in range(N+1)]
    visited[idx]=True
    retrunKevin=0
    while q:
        current,kevinBacon=q.popleft()
        retrunKevin+=kevinBacon
        if current in dic:
            for i in dic[current]:
                if not visited[i]:
                    visited[i]=True
                    q.append((i,kevinBacon+1))
    kevinList[idx]=retrunKevin

for i in range(1,N+1):
    bfs(i)

targetBaconNum=min(kevinList[1:])
for i in range(1,N+1):
    if kevinList[i]==targetBaconNum:
        print(i)
        break
```

# 백준 1654 랜선 자르기 실버 2

```python
K, N = map(int,input().split())
lanList=[]
for i in range(K):
    lanList.append(int(input()))
lanList.sort()

s=1
e=max(lanList)

while s<=e:
    cutLanNum=0
    mid=(s+e)//2
    for i in lanList:
        cutLanNum+=i//mid
    if cutLanNum>=N:
        s=mid+1
    else:
        e=mid-1 

print(e)
```

# 백준 2110 공유기 설치 골드 4

```python
N, C = map(int,input().split())
li=[int(input()) for _ in range(N)]
li.sort()

s, e = 1, li[-1] - li[0] 
result = 0 

while s <= e:
    mid = (s + e) // 2  
    count = 1 
    last_pos = li[0]

    for i in range(1, N):
        if li[i] >= last_pos + mid:
            count += 1 
            last_pos = li[i] 
    if count >= C:
        result = mid 
        s = mid + 1
    else:
        e = mid - 1

print(result)
```
