# 백준 1697 숨바꼭질 실버 1

```python
from collections import deque
N,K=map(int,input().split())
Max=100001
visited=[False for _ in range(Max)]
q=deque()
q.append((N,0))
visited[N]=True
while q:
    pos,time=q.popleft()
    if pos==K:
        print(time)
        break
    for p in (pos*2,pos+1,pos-1):
        if 0<=p<Max and not visited[p]:
            visited[p]=True
            q.append((p,time+1))
```

# 백준 12851 숨바꼭질 2 골드 4

```python
from collections import deque
N,K=map(int,input().split())
Max=100001
visited=[float('inf') for _ in range(Max)]
q=deque([(N,0)])
minTime=float('inf')
cnt=0
while q:
    p,t=q.popleft()
    if p==K:
        if minTime>t:
            cnt=0
            minTime=t
        cnt+=1
    for pos in (p*2,p+1,p-1):
        if 0<=pos<Max and t+1<=visited[pos]:
            visited[pos]=t+1
            q.append((pos,t+1))

if N==K:
    minTime=0
    cnt=1
print(minTime,cnt,sep='\n')

```

# 백준 1912 연속합 실버2

```python
N=int(input())
arr=list(map(int,input().split()))

maxN=float('inf')*-1
sumNum=arr[0]

next=0
def ham(n):
    global next
    if i<N:
        for i in range(n,N):
            if sumNum>maxN:
                maxN=sumNum
                next=i
            sumNum+=arr[i]
        next=i
    next = None

ham(1)
while 1:
    if next==None:
        break
    else:
        ham(next)

print(maxN)
```

# 백준 1600 말이 되고픈 원숭이 골드3

```python
from collections import deque
import sys
input=sys.stdin.readline
K=int(input())
W,H=map(int,input().split())
arr=[]
q=deque([(0,0,K,0)])
visited=[[[False for _ in range(W)] for _ in range(H)] for _ in range(K+1)]
visited[K][0][0]=True
#x,y순
knightMove=[(-2,1),(-1,2),(1,2),(2,1),(-2,-1),(-1,-2),(1,-2),(2,-1)]
monkeyMove=[(1,0),(-1,0),(0,1),(0,-1)]
for _ in range(H):
    line=list(map(int,input().split()))
    arr.append(line)

while q:
    hereX,hereY,k,moveNum=q.popleft()
    
    if hereY==H-1 and hereX==W-1:
        print(moveNum)
        exit()
    if k>0:
        #나이트 움직임 하나하나 대입
        for move in knightMove:
            moveX=move[0]
            moveY=move[1]
            #가능한가 체크
            if 0<=hereY+moveY<H and 0<=hereX+moveX<W and arr[hereY+moveY][hereX+moveX]!=1 and not visited[k-1][moveY+hereY][moveX+hereX]:
                visited[k-1][moveY+hereY][moveX+hereX]=True
                q.append((hereX+moveX,hereY+moveY,k-1,moveNum+1))
    for move in monkeyMove:
        moveX=move[0]
        moveY=move[1]
        if 0<=hereY+moveY<H and 0<=hereX+moveX<W and arr[hereY+moveY][hereX+moveX]!=1 and not visited[k][moveY+hereY][moveX+hereX]:
            visited[k][moveY+hereY][moveX+hereX]=True
            q.append((hereX+moveX,hereY+moveY,k,moveNum+1))

print(-1)

```

# 백준 2146번 다리 만들기 골드3

```python
from collections import deque
import sys

# 입력 속도 개선
input = sys.stdin.readline

N = int(input())
arr = []
for _ in range(N):
    arr.append(list(map(int, input().split())))

visited = [[False] * N for _ in range(N)]
dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]
island_num = 1

# 1단계: BFS로 각 섬에 고유 번호 붙이기 (2, 3, 4...)
def number_islands(x, y):
    global island_num
    q = deque([(x, y)])
    visited[x][y] = True
    arr[x][y] = island_num
    
    while q:
        cur_x, cur_y = q.popleft()
        for i in range(4):
            nx, ny = cur_x + dx[i], cur_y + dy[i]
            if 0 <= nx < N and 0 <= ny < N and not visited[nx][ny] and arr[nx][ny] == 1:
                visited[nx][ny] = True
                arr[nx][ny] = island_num
                q.append((nx, ny))

for i in range(N):
    for j in range(N):
        if arr[i][j] == 1 and not visited[i][j]:
            number_islands(i, j)
            island_num += 1

# 2단계: "만남의 장소" 방식으로 최단 다리 찾기
dist = [[-1] * N for _ in range(N)]
q = deque()

# 모든 육지 칸을 시작점으로 큐에 추가
for i in range(N):
    for j in range(N):
        if arr[i][j] > 0:
            q.append((i, j))
            dist[i][j] = 0

min_bridge = float('inf')

while q:
    x, y = q.popleft()
    
    for i in range(4):
        nx, ny = x + dx[i], y + dy[i]
        
        if 0 <= nx < N and 0 <= ny < N:
            # 다른 섬의 영토를 만난 경우 (다리 완성)
            if arr[nx][ny] > 0 and arr[nx][ny] != arr[x][y]:
                bridge_len = dist[x][y] + dist[nx][ny]
                min_bridge = min(min_bridge, bridge_len)
            
            # 주인이 없는 바다를 만난 경우 (영토 확장)
            if arr[nx][ny] == 0 and dist[nx][ny] == -1:
                dist[nx][ny] = dist[x][y] + 1
                arr[nx][ny] = arr[x][y] # 바다 채워버려
                q.append((nx, ny))

print(min_bridge)
```

# 백준 10971 외판원 순회 2

```python
from collections import deque
N=int(input())
arr=[]
visited=[False for _ in range(N+1)]
g={}
for _ in range(N):
    line=list(map(int,input().split()))
    arr.append(line)

for i in range(N):
    for j in range(N):
        if arr[i][j]!=0:
            g.setdefault(i,[]).append((j,arr[i][j]))

min_cost=float('inf')
def dfs(start_city,current_city,count,current_cost):
    global min_cost
    if current_cost>=min_cost:
        return
    if count == N:
        if arr[current_city][start_city]!=0:
            total_cost=current_cost+arr[current_city][start_city]
            min_cost=min(min_cost,total_cost)
        return
    for city in g[current_city]:
        if arr[current_city][city[0]]!=0 and not visited[city[0]]:
            visited[city[0]]=True
            dfs(city[0],count+1,current_cost+city[1])
            visited[city[0]]=False

for city in g:
    visited[city]=True
    dfs(city,city,1,0)
    visited[city]=False

print(min_cost)
```

# 백준 1987 알파벳 골드 4

```python
from collections import deque

R, C = map(int, input().split())
arr = [list(input()) for _ in range(R)]

dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]

start = 1 << (ord(arr[0][0]) - ord('A'))

visited = set()
visited.add((0, 0, start))

q = deque()
q.append((0, 0, start, 1))

mxcount = 1

while q:
    x, y, bitmask, count = q.popleft()
    mxcount = max(mxcount, count)

    for i in range(4):
        nx, ny = x + dx[i], y + dy[i]
        if 0 <= nx < C and 0 <= ny < R:
            c = ord(arr[ny][nx]) - ord('A')
            if not (bitmask & (1 << c)):
                nbitmask = bitmask | (1 << c)
                state = (ny, nx, nbitmask)
                if state not in visited:
                    visited.add(state)
                    q.append((nx, ny, nbitmask, count + 1))

print(mxcount)
```
